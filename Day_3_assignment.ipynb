```python
import seaborn as sns 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fmri=sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',data=fmri,hue='subject')
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri,hue='subject')
```

```python
sns.catplot(x='timepoint',y='region',kind='box',data=fmri)
```

```python
sns.catplot(x='timepoint',y='region',kind='box',data=fmri,hue='subject')
```
